```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
deathsByState = pd.read_csv("Datasets\Provisional_COVID-19_Death_Counts_in_the_United_States_by_County.csv")
```

```python
#rename columns to match
deathsByState = deathsByState.rename(columns={"FIPS County Code": "FIPS", "County name": "County"})
```

```python
#drops rows with no covid deaths
deathsByState.dropna(subset=['Deaths involving COVID-19'], inplace=True)
deathsByState
```

```python
#drop unnecessary columns
deathsByState.drop(['Footnote', 'Deaths from All Causes'], axis=1, inplace=True)
deathsByState
```

```python
#group by state, county, fips, urabn rural code, and sum deaths
deathsByState = deathsByState.groupby(['State', 'County', 'FIPS', 'Urban Rural Code'])['Deaths involving COVID-19'].sum()
deathsByState
```

```python
vaccsByCounty = pd.read_csv("Datasets\COVID-19_Vaccinations_in_the_United_States_County.csv")
```

```python
vaxPerCounty = pd.DataFrame ({
    "Date": vaccsByCounty['Date'],
    "FIPS": vaccsByCounty['FIPS'],
    "County": vaccsByCounty['Recip_County'],
    "State": vaccsByCounty['Recip_State'],
    "Total_Vaccinated": vaccsByCounty['Series_Complete_Yes']
})

vaxPerCounty
```

```python
#drop rows with a date value that isnt 01/25/2023 or later
vaxPerCounty = vaxPerCounty[(vaxPerCounty[['Date']] == '01/25/2023').all(axis=1)]
vaxPerCounty
```

```python
#drop rows with 0 vaccines
vaxPerCounty = vaxPerCounty[(vaxPerCounty[['Total_Vaccinated']] != 0).all(axis=1)]
vaxPerCounty.dropna(subset=['Total_Vaccinated'], inplace=True)
vaxPerCounty
```

```python
#sort alphabetically by county
vaxPerCounty.sort_values(by=['State'], inplace=True)
vaxPerCounty
```

```python
#drop rows with UNK fips
vaxPerCounty = vaxPerCounty[vaxPerCounty.FIPS != 'UNK']
vaxPerCounty
```

```python
#group by total vaccinated and sum
vaxPerCounty = vaxPerCounty.groupby(['County', 'State', 'FIPS'])['Total_Vaccinated'].sum()
vaxPerCounty
```

```python
#merge dataframes
merged = pd.merge(vaxPerCounty, deathsByState, on=['County', 'State'])
merged.reset_index(inplace=True)
merged
```

```python
#remove county, parish and scity from county names
merged["County"] = merged["County"].str.replace("\sCounty", "")
merged["County"] = merged["County"].str.replace("\sParish", "")
merged["County"] = merged["County"].str.replace("\scity", "")
```

```python
#merge with county incomes by county
countyIncome = pd.read_csv("Datasets\countyincome.csv")
merged2 = pd.merge(merged, countyIncome, on=['County'])
merged2
```

```python

#get rid of repeat counties
merged2.drop_duplicates(subset=['County', 'State'], keep='first', inplace=True)
merged2
```

```python
#make csv
merged2.to_csv('Datasets\countyVaccsVSCountyDeaths.csv', index=False)
```
