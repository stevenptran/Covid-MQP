```python
import pandas as pd
from sodapy import Socrata
```

```python
client = Socrata("data.cdc.gov", None)
results = client.get("q3t8-zr7t", limit=1000000)

df = pd.DataFrame.from_records(results)
df
```

```python
df = df[['group', 'subgroup', 'value', 'start_time','measure']]
df = df.rename(columns={'group': 'Group',
                        'subgroup': 'Subgroup',
                        'value': 'Value',
                        'start_time': 'Start_Time',
                        'measure': 'Measure'})
```

```python
df = df[df['Measure'] == 'Average length of stay (days)']
```

```python
df1 = df[df['Group'] == 'Discharge Status']
df = df[df['Group'] == 'Age']
```

```python
df = df.reset_index()
df = df.drop('index', axis=1)
df = df.drop('Measure', axis=1)

df1 = df1.reset_index()
df1 = df1.drop('index', axis=1)
df1 = df1.drop('Measure', axis=1)
```

```python
df
df.to_csv('Datasets/AvgLengthOfStaysAge.csv')
```
