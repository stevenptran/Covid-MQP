```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as npy
import datetime as dt
import matplotlib.dates as mdates

Provisional_COVID_Deaths = pd.read_csv("Datasets/Provisional_COVID-19_Deaths_by_Sex_and_Age.csv")
prov = pd.DataFrame(Provisional_COVID_Deaths)
State_Populations = pd.read_csv("Datasets/StatePops.csv")
pops = pd.DataFrame(State_Populations)

#source: https://worldpopulationreview.com/state-rankings/state-densities
State_Densities = pd.read_csv("Datasets/StatePopDense.csv")
dens = pd.DataFrame(State_Densities)
#print(prov)
```

```python
#select only All Ages, All Sexes, and By Total in specified columns
all_ages = prov.loc[(prov['Age Group'] == 'All Ages') & (prov['Sex'] == 'All Sexes') & (prov['Group'] == 'By Total')]
#all_ages
```

```python
#drop nonstates
all_ages.drop(all_ages[all_ages['State'] == 'United States'].index, inplace = True)
all_ages.drop(all_ages[all_ages['State'] == 'District of Columbia'].index, inplace = True)
all_ages.drop(all_ages[all_ages['State'] == 'Puerto Rico'].index, inplace = True)

all_ages = all_ages.replace('New York City', 'New York')

unique = pd.unique(all_ages['State'])

#print(len(unique))
#print(unique)
```

```python
#reset indices
all_ages.reset_index(drop=True, inplace=True)
#all_ages
```

```python
#remove unused columns leaving only Data As Of, State, COVID-19 Deaths, and Total Deaths
cleaned2 = all_ages[['Data As Of', 'State', 'COVID-19 Deaths', 'Total Deaths']]
cleaned2
```

```python
#select only state and 2022 population from state populations dataset
pops2 = pops[['state', 'pop2022']]
pops2
```

```python
dens2 = dens[['state', 'pop2022', 'densityMi']]
dens2
```

```python
#merge covid deaths dataset with state populations dataset
new = pd.merge(cleaned2, dens2, how='inner', left_on = 'State', right_on = 'state')
new.drop(columns=['state'], inplace=True)
new = new.groupby('State').sum()
new
```

```python
with open('covdeaths_statepops.txt', 'w') as csv_file:
    new.to_csv(path_or_buf=csv_file)
```
