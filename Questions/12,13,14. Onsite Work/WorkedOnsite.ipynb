```python
import pandas as pd
import numpy as npy
from pathlib import Path  

WorkedOnSite = pd.read_csv("Datasets/WorkedOnsiteWeeks34-45.csv")
WorkedOnSite
```

```python
#Isolating US and cut the front 2 numbers of each date to get just the week number
WorkedOnSite = WorkedOnSite.iloc[3:]
USwos = WorkedOnSite.iloc[:12]
USwos['Unnamed: 1'] = USwos['Unnamed: 1'].str[:2]
#USwos
filepath = Path('Datasets/USOnsite.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
USwos.to_csv(filepath)  

```

```python
#Isolating Week45 States
Statewos45 = WorkedOnSite.iloc[13:]
Statewos45 = Statewos45.iloc[:51]
Statewos45.drop([24], inplace=True)
Statewos45

filepath = Path('Datasets/StateOnsite45.csv')      
filepath.parent.mkdir(parents=True, exist_ok=True)  
Statewos45.to_csv(filepath)  
```

```python
#Isolating Week34 States
Statewos34 = WorkedOnSite.iloc[574:]
Statewos34 = Statewos34.iloc[:51]
Statewos34.drop([585], inplace=True)
Statewos34
  
filepath = Path('Datasets/StateOnsite34.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
Statewos34.to_csv(filepath)  
```
