```python
import pandas as pd
import datetime as dt
```

```python
restaurantDf = pd.read_csv(r"Datasets\U.S._State_and_Territorial_Orders_Closing_and_Reopening_Restaurants_Issued_from_March_11__2020_through_August_15__2021_by_County_by_Day.csv")
```

```python
#restaurantDf
restaurantDf = restaurantDf[restaurantDf.State_Tribe_Territory != "DC"]
restaurantDf = restaurantDf[restaurantDf.State_Tribe_Territory != "MP"]
restaurantDf = restaurantDf[restaurantDf.State_Tribe_Territory != "AS"]
restaurantDf = restaurantDf[restaurantDf.State_Tribe_Territory != "GU"]
restaurantDf = restaurantDf[restaurantDf.State_Tribe_Territory != "PR"]
restaurantDf = restaurantDf[restaurantDf.State_Tribe_Territory != "VI"]

#restaurantDf.State_Tribe_Territory.unique()
#len(restaurantDf.State_Tribe_Territory.unique())
restaurantDf = restaurantDf.rename(columns = {"State_Tribe_Territory":"state"})
```

```python
stateConverter = pd.read_csv("Datasets\StateConverter.csv")

stateConverter
```

```python
restaurantDf = restaurantDf.merge(stateConverter, on="state")
#restaurantDf.to_csv("50statetest.csv")
```

```python
restaurantDf = restaurantDf.drop(['state'], axis=1)
restaurantDf = restaurantDf.drop(['Abbrev'], axis=1)
restaurantDf['date'] = pd.to_datetime(restaurantDf['date'], format='%m/%d/%Y')
```

```python
#restaurantDf.head()

restaurantDf.Action.unique()
restaurantDf['date'] = (restaurantDf['date'] - dt.datetime(1970,1,1)).dt.total_seconds()
restaurantDf
```

```python
dates = {'Date':['2020-03-01', '2020-04-01', '2021-03-01', '2021-04-01']}
dateConverter = pd.DataFrame(dates)
dateConverter['Date'] = pd.to_datetime(dateConverter['Date'], format='%Y-%m-%d')
dateConverter['Epoch'] = (dateConverter['Date'] - dt.datetime(1970,1,1)).dt.total_seconds()
#dateConverter
```

```python
restaurantDf['march2020'] = restaurantDf['date'].between(dateConverter['Epoch'][0], dateConverter['Epoch'][1], inclusive=True)
restaurantDf['march2021'] = restaurantDf['date'].between(dateConverter['Epoch'][2], dateConverter['Epoch'][3], inclusive=True)

mar2020df = restaurantDf[ restaurantDf['march2020'] != False ]
mar2021df = restaurantDf[ restaurantDf['march2021'] != False ]

mar2020df
mar2021df
```

```python
actionsDf2020 = mar2020df.groupby(['State', 'Action'])['Action'].count()
actionsDf2021 = mar2021df.groupby(['State', 'Action'])['Action'].count()
```

```python
actionsDf2020 = actionsDf2020.unstack()
actionsDf2020['Totals'] = actionsDf2020.sum(axis=1)

actionsDf2021 = actionsDf2021.unstack()
actionsDf2021['Totals'] = actionsDf2021.sum(axis=1)
```

```python
actionsDf2020 = actionsDf2020.iloc[:,:-1].div(actionsDf2020.Totals, axis=0)
actionsDf2021 = actionsDf2021.iloc[:,:-1].div(actionsDf2021.Totals, axis=0)
```

```python
actionsDf2020 = actionsDf2020.fillna(0)
actionsDf2020
```
