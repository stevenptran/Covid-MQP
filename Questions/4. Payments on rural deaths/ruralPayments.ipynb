```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ruralPayments = pd.read_csv('Datasets\American_Rescue_Plan__ARP__Rural_Payments.csv')   
ruralPayments.head()
```

```python
#remove the $ and , from the columns
ruralPayments['Total Payments'] = ruralPayments['Payment'].str.replace('$','')
ruralPayments['Total Payments'] = ruralPayments['Total Payments'].str.replace(',','')
ruralPayments['Total Payments'] = ruralPayments['Total Payments'].str.replace(' ','')
ruralPayments['Total Payments'] = ruralPayments['Total Payments'].astype(float)
ruralPayments['Total Payments'] = ruralPayments['Total Payments'].astype(int)
ruralPayments
```

```python
#group by state and rural region and sum the payments
ruralPayments = ruralPayments.groupby(['State', 'Rural Location'])['Total Payments'].sum()
ruralPayments
```

```python
#convert to a dataframe
ruralPayments = ruralPayments.to_frame()
ruralPayments
```

```python
#convert to a csv
ruralPayments.to_csv('Datasets\ARP_Rural_Payments.csv')
```
