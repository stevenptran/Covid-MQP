```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
placeOfDeath = pd.read_csv('Datasets\AH_Cumulative_Provisional_COVID-19_Death_Counts_by_Place_of_Death_and_Age_Group_from_2_1_2020_to_7_18_2020.csv')
placeOfDeath
```

```python
#drop unnecessary columns
placeOfDeath = placeOfDeath.drop(['HHS Region', 'Deaths involving Pneumonia, with or without COVID-19, excluding Influenza deaths\n(J12.0–J18.9)', 'Deaths involving COVID-19 and Pneumonia, excluding Influenza (U07.1 and J12.0–J18.9)'], axis=1)
placeOfDeath
```

```python
#group state, place of death, age group and sum all deaths involving covid-19
placeOfDeath = placeOfDeath.groupby(['State', 'Place of Death', 'Age group'])['All Deaths involving COVID-19 (U07.1)'].sum()
placeOfDeath
```

```python
#convert to dataframe
placeOfDeath = pd.DataFrame(placeOfDeath)
placeOfDeath
```

```python
#normalize data
placeOfDeath = placeOfDeath.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))
placeOfDeath
```

```python
#convert to csv 
placeOfDeath.to_csv('Datasets\placeOfDeath.csv')
```
