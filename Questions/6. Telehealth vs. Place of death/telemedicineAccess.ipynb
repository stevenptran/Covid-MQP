```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
teleAccess = pd.read_csv('Datasets\Access_and_Use_of_Telemedicine_During_COVID-19.csv')
teleAccess.head()
```

```python
#drop columns that are not needed
teleAccess.drop(['Round', 'Standard Error', 'Suppression', 'Significant 1', 'Significant 2'], axis=1, inplace=True)
teleAccess
```

```python
#convert the data to numeric
teleAccess['Percent'] = pd.to_numeric(teleAccess['Percent'], errors='coerce')
teleAccess
```

```python
#convert to csv
teleAccess.to_csv('Datasets\\teleAccess.csv', index=False)
```

```python
#convert to json
teleAccess.to_json('Datasets\\teleAccess.json', orient='records')
```
