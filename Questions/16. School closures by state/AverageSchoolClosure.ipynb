```python
import pandas as pd
import datetime
from sodapy import Socrata
```

```python
client = Socrata("data.cdc.gov", None)
results = client.get("wgvr-7mvz", limit=1000000)


df = pd.DataFrame.from_records(results)
df
```

```python
dateClosure = df['dateclosure'].str.slice(stop=10)
df['dtClosure'] = pd.to_datetime(dateClosure, format='%Y/%m/%d')
df['epochClosure'] = df['dtClosure'].apply(lambda x: int(x.timestamp()))
df
```

```python
dfmeans = df.groupby(['state']).mean()
dfmeans
```

```python
dfmeans['ClosureDate'] = pd.to_datetime(dfmeans['epochClosure'], unit='s')
dfmeans = dfmeans.sort_values(by=['epochClosure'])
dfmeans = dfmeans.reset_index()
dfFinal = dfmeans[['state', 'ClosureDate']]
dfFinal['ClosureDate'] = dfFinal['ClosureDate'].apply(lambda x: x.strftime('%Y-%m-%d'))
dfFinal
```

```python
dfFinal.to_csv('AverageSchoolClosure.csv')
```
