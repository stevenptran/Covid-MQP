```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.bls.gov/lau/lastrk20.htm"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')

table = soup.find('table', {'id': 'lastrk20'})
```

```python
headers = [th.text.strip() for th in table.find_all('th')][:3]
```

```python
data = []
flag = True
for row in table.find_all('tr'):
    if flag:
        flag = False
        continue
    cols = row.find_all(['th', 'td'])
    cols = [col.text.strip() for col in cols]
    if len(cols) > 1:
        data.append([col for col in cols if col])
```

```python
#print(headers)
print(data)
df = pd.DataFrame(data, columns=headers).drop('Rank', axis=1)
```

```python
df = df.drop(df[(df['State'] == 'United States') | (df['State'] == 'District of Columbia')].index)
df
```

```python
df.to_csv('Datasets/unemployment_rate2020.csv', index=False)
```
