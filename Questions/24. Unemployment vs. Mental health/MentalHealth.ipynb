```python
import pandas as pd
```

```python
df = pd.read_csv('Datasets\Mental_Health_Care_in_the_Last_4_Weeks.csv')
df
```

```python
df = df[['Indicator', 'Group', 'State', 'Subgroup', 'Time Period Start Date','Value']]
df
```

```python
df = df[df['Group'] == 'By State']
df
```

```python
df = df[df['State'] != 'United States']
df
```

```python
df = df.drop('Subgroup', axis=1)
df = df.drop('Group', axis=1)
```

```python
df = df.reset_index()
df = df.drop('index', axis=1)
```

```python
df.to_csv('Datasets/MentalHealth.csv')
```
