```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
healthInsurance = pd.read_csv('Datasets\Indicators_of_Health_Insurance_Coverage_at_the_Time_of_Interview.csv')
```

```python
#drop rows with missing values
healthInsurance.dropna(subset=['Value'], inplace=True)
healthInsurance
```

```python
#drop unnecessary columns
healthInsurance.drop(['Low CI', 'High CI', 'Confidence Interval', 'Quartile Range', 'Quartile Number', 
'Suppression Flag'], axis=1, inplace=True)
healthInsurance
```

```python
#group by state, year, indicator, phase, and value
healthInsurance = healthInsurance.groupby(['State', 'Indicator'])['Value'].sum()
healthInsurance
```

```python
#remove united states
healthInsurance = healthInsurance.drop('United States')
healthInsurance
```

```python
#unstack data
healthInsurance = healthInsurance.unstack()
healthInsurance
```

```python
#normalize data
healthInsurance = healthInsurance.div(healthInsurance.sum(axis=1), axis=0)
healthInsurance
```

```python
#convert to csv
healthInsurance.to_csv('Datasets\healthInsurance.csv')
```
