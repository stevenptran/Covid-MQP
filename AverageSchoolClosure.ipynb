```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('Datasets/COVID-19-associated_school_closures__United_States__February_18_June_30__2020.csv')
df
```

```python
dateClosure = df['DateClosure'].str.slice(stop=10)
df['dtClosure'] = pd.to_datetime(dateClosure, format='%m/%d/%Y')
#df['epochClosure'] = df['dtClosure'].apply(lambda x: x.timestamp())
df['epochClosure'] = df['dtClosure'].astype(int)
df
```

```python
dfmeans = df.groupby(['State']).mean()
dfmeans
```

```python
dfmeans['ClosureDate'] = pd.to_datetime(dfmeans['epochClosure'])
dfmeans = dfmeans.sort_values(by=['epochClosure'])
dfmeans = dfmeans.reset_index()
dfFinal = dfmeans[['State', 'ClosureDate']]
dfFinal
```

```python
dfFinal.to_csv('AverageSchoolClosure.csv')
```
